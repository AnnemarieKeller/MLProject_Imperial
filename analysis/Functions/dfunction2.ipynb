```python
import numpy as np
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler



# Standardize inputs (helps SVR)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)

# SVR ensemble for pseudo-uncertainty
n_ensembles = 5
svr_ensemble = [SVR(kernel='rbf', C=10.0, epsilon=0.01) for _ in range(n_ensembles)]

# Fit ensemble
for model in svr_ensemble:
    model.fit(X_train_scaled, y_train)

input_dim = X_train.shape[1]
num_iterations = 30
num_candidates = 1000

for i in range(num_iterations):
    # Generate random candidates
    candidates = np.random.rand(num_candidates, input_dim)
    candidates_scaled = scaler.transform(candidates)

    # Predict with ensemble
    preds = np.array([m.predict(candidates_scaled) for m in svr_ensemble])
    mean_pred = preds.mean(axis=0)
    std_pred = preds.std(axis=0)

    # Upper Confidence Bound acquisition
    beta = 2.0  # exploration/exploitation trade-off
    ucb = mean_pred + beta * std_pred

    # Pick next candidate
    idx_next = np.argmax(ucb)
    next_input = candidates[idx_next]
    next_output = np.sin(np.sum(next_input))  # Replace with your real function

    print(f"Iter {i+1}: Next input = {next_input}, Predicted output = {mean_pred[idx_next]:.4f}, UCB = {ucb[idx_next]:.4f}, Real output = {next_output:.4f}")

    # Add new observation to training set
    X_train = np.vstack([X_train, next_input])
    y_train = np.append(y_train, next_output)
    X_train_scaled = scaler.fit_transform(X_train)

    # Refit ensemble
    svr_ensemble = [SVR(kernel='rbf', C=10.0, epsilon=0.01) for _ in range(n_ensembles)]
    for model in svr_ensemble:
        model.fit(X_train_scaled, y_train)

# After loop, find best observed
best_idx = np.argmax(y_train)
print("\nBest input found:", X_train[best_idx])
print("Best output observed:", y_train[best_idx])
```
